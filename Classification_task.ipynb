```python
# using preprocessing approach to labelencoder
Encoder_dataset=dataset.copy()
from sklearn.preprocessing import LabelEncoder
label_en=LabelEncoder()
Encoder_dataset=Encoder_dataset.apply(label_en.fit_transform)
Encoder_dataset
```

# Classifying the class of the mushroom using LogisticRegression

```python
#classifying the class of the mushroom using logistic regression
X=Encoder_dataset.drop('class',axis=1)
y=Encoder_dataset['class']
```

```python
from sklearn.linear_model import LogisticRegression
logistic_reg=LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,stratify=y,test_size=0.3)
```

```python
logistic_reg.fit(X_train,y_train)
```

```python
y_predict=logistic_reg.predict(X_test)
y_predict
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_predict,y_test)
```

# Creating classification report and geneating confusion matrix of the prediceted class

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
# Generating confusion matrix 
confus_matrix=confusion_matrix(y_test,y_predict)
print(confus_matrix)
```

```python
import seaborn as sb
sb.heatmap(confus_matrix,annot=True)
```

```python
#creatin classification report
report=classification_report(y_test,y_predict)
print(report)
```
